```python
from glob import glob
import pyterrier as pt
from tira.third_party_integrations import ensure_pyterrier_is_loaded
from pathlib import Path
```

```python
ensure_pyterrier_is_loaded()
ds_id = "radboud-validation-20251114-training"
#dataset = pt.datasets.get_dataset(f"irds:ir-lab-wise-2025/{ds_id}")
dataset = pt.datasets.get_dataset(f"irds:ir-lab-wise-2025/radboud-validation-20251114-training")
topics = dataset.get_topics("title")
qrels = dataset.get_qrels()
```

```python
runs = {}
#run_name = f'output/runs/{ds_id}/*/run.txt.gz'
#run_name = f'best-try/runs-2/{ds_id}/*/run.txt.gz'
#run_name = f"Test/run/{ds_id}/*/run.txt.gz"
run_name = "Test/runs/radboud-validation-20251114-training"
#for run_name in glob(f'output/runs/{ds_id}/*/run.txt.gz'):
for run_name in Path(run_name).iterdir():
    i = 1
    run_name = f"{run_name}/run.txt.gz"
    print(f"Loaded run: {run_name}")
    runs[run_name.split('/')[-2]] = pt.io.read_results(run_name)
    i += 1

pt.Experiment(
    list(runs.values()),
    topics,
    qrels,
    ["ndcg_cut.10", "map", "P_10", "recip_rank"],
    list(runs.keys())
)
    
```
